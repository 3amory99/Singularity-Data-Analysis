## Apply Functions

```python
df = pd.DataFrame({'col1':[1,6,14,5,11,3,2,4],'col2':[123,456,789,101112,131415,161718,192021,222324],'col3':['abc','def','ghi','jkl', 'opq','rst','uvw','xyz']})
df
```

```python
df['extention'] = df['col1'] + 10
df
```

```python
df['radius'] = df['col1'] + 5
df
```

```python
def area(r):
    return np.pi * (r ** 2)
```

```python
df['area'] = df['radius'].apply(area)
df
```

```python
df['lambda_expression'] = df['radius'].apply(lambda r: np.pi * (r ** 2))
df
```
