# Group By
- **Allows you to group rows of data together and you can also call aggregate functions.**

```python
import pandas as pd
import numpy as np
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('covid_19_data.csv')
df.head(10)
```

### Groupwise analysis

One function we've been using heavily thus far is the value_counts() function. We can replicate what value_counts() does by doing the following:

```python
df.groupby('Recovered').mean()
```

```python
df.groupby('Confirmed').Confirmed.count()
```

```python
df.groupby('Province/State').Deaths.min()
```

```python
df.groupby('Province/State').Confirmed.std()
```

```python
df2 = pd.read_csv('albb-salaries-2003.csv')
df2.head(10)
```

```python
df2.groupby('Salary').First.max()
```

```python
df2.groupby('Salary').Position.max()
```

```python
df2.groupby('Salary').Position.max().head(3)
```

```python
df2.groupby('Salary').Position.max().unique()
```

```python
df2
```

```python
df2.groupby(['Team']).Salary.agg([len,max,min])
```

```python
df2.reset_index()
```

```python
df2.groupby(['Salary']).mean().sort_values('Salary' , ascending = False).head(10)
```
