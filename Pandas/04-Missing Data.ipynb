# Missing Data

```python
# Import libs
import numpy as np
import pandas as pd
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
nums = np.random.rand(7,7)
df = pd.DataFrame(nums, 
                  index = ['2021-01-20','2021-01-21','2021-01-22','2021-01-23','2021-01-24','2021-01-25','2021-01-26'],
                  columns = ['C1','C2','C3','C4','C5','C6','C7'])
```

```python
df
```

```python
df.at[:, 'C8'] = np.NaN
df.at[0:2 , 'C6'] = np.NaN
df.at[5:7 , 'C5'] = np.NaN
df.at[1:4 , 'C2'] = np.NaN
```

```python
df
```

```python
# Detect Missing Values or NULL Values
df.isna()
```

```python
# Detect Non-Missing Values
df.notna()
```

### Fill missing data

```python
# Fill all NULL values with 1020
df.fillna(0)
```

```python
df.fillna("Not Null")
```

```python
# Replace Null values in Column 'C2' with number 12345
# Replace Null values in Column 'C8' with number 6789
# Replace Null values in Column 'C5' with number 15432
df.fillna(value={'C2' : 12345 ,'C5' : 15432, 'C8' : 6789})
```

```python
#Replace first NULL value in Column C8 with 6789
df.fillna(value={'C8' : 6789} , limit=1)
```

```python
#Replace all NULL value except last value in Column C8 with 6789
df.fillna(value={'C8' : 6789} , limit = len(df.C8)-1)
```

### Drop missing data

```python
# Drop Rows with NULL values
df.loc['2021-01-26','C5'] = 44
df.drop('C8', axis=1,inplace=True)
df
```

```python
# Drop Rows with NULL values
# Row is a defult parameter
df.dropna()
```

```python
# Drop Columns with NULL values
df.dropna(axis='columns')
```

```python
df
```

```python
# Drop Rows with NULL values present in C2 or C5
df.dropna(subset=['C2' ,'C5'])
```

```python
df.loc["2021-01-27"] =[np.NAN,np.NAN,np.NAN,np.NAN,np.NAN,np.NAN,0.05052]
df
```

```python
# At least return one value through rows dimentions
df.dropna(thresh=1, axis=0)
```

```python
# At least return one value through rows dimentions
df.dropna(thresh=2, axis=0)
```

```python
df.dropna(thresh=3, axis=0)
```

```python
# At least return one value through rows dimentions
df.dropna(thresh=7, axis=0)
```

```python
# Fall all Null values of mean of its column
for col in df.columns:
    print(df[col].mean())
    df[col].fillna(value=df[col].mean(), inplace=True)
df
```
