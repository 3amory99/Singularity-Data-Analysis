## Importing multiple CSV files in DataFrame


```python
import glob
```

```python
# Append all CSV files 
path =r''
filenames = glob.glob(path + "/*.csv")
covid = pd.DataFrame()
for f in filenames:
    df = pd.read_csv(f)
    covid = covid.append(df,ignore_index=True,sort=True)
```
